# The KÃ¼nigsberg Bridge Problem

https://en.wikipedia.org/wiki/Seven_Bridges_of_K%C3%B6nigsberg

Can we visualize this probelm in python?

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()


G.add_edge(1,2)
G.add_nodes_from([1,2,3,4,5])

G.add_edges_from([(1,2,{'color':'blue'}), (2,3,{'weight':8})])

nx.draw(G)
plt.show()
```

```python
K = nx.MultiGraph()
K.add_nodes_from(['north', 'south', 'east', 'west'])
K.add_edges_from([('west', 'north'), ('west', 'north'),
                  ('west', 'south'), ('west', 'south')])
K.add_edges_from([('east', 'north'), ('east', 'south')])
K.add_edge('east', 'west')
nx.draw(K)
plt.show()
```
